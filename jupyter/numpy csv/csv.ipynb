```python
import numpy as np
import matplotlib.pyplot as plt
import csv

# Fungsi untuk membaca file CSV ke dalam array NumPy
def csv_to_numpy(filename):
    data = np.genfromtxt(filename, delimiter=',', skip_header=1, invalid_raise=False)
    return data

# Fungsi untuk analisis data (rata-rata, max, min)
def analyze_data(data):
    if data is None:
        print("No data to analyze")
        return None

    # Membuat array indeks untuk tanggal (karena data tanggal dalam format numerik bisa tidak valid)
    dates = np.arange(len(data))  
    
    # Ambil kolom yang relevan dari data
    temperature = data[:, 1]  # Kolom temperature
    humidity = data[:, 2]     # Kolom humidity
    wind_speed = data[:, 3]   # Kolom wind_speed

    print("Temperature Analysis:")
    print("Mean Temperature:", np.nanmean(temperature))  # Menangani NaN
    print("Max Temperature:", np.nanmax(temperature))    # Menangani NaN
    print("Min Temperature:", np.nanmin(temperature))    # Menangani NaN

    print("\nHumidity Analysis:")
    print("Mean Humidity:", np.nanmean(humidity))  # Menangani NaN
    print("Max Humidity:", np.nanmax(humidity))    # Menangani NaN
    print("Min Humidity:", np.nanmin(humidity))    # Menangani NaN

    print("\nWind Speed Analysis:")
    print("Mean Wind Speed:", np.nanmean(wind_speed))  # Menangani NaN
    print("Max Wind Speed:", np.nanmax(wind_speed))    # Menangani NaN
    print("Min Wind Speed:", np.nanmin(wind_speed))    # Menangani NaN

    return dates, temperature, humidity, wind_speed  # Pastikan ini ada!

# Fungsi untuk visualisasi data
def plot_data(dates, temperature, humidity, wind_speed):
    plt.figure(figsize=(12, 8))

    # Plot Temperature
    plt.subplot(3, 1, 1)
    plt.plot(dates, temperature, color='r', marker='o', linestyle='-', label='Temperature')
    plt.xlabel('Index')
    plt.ylabel('Temperature (C)')
    plt.title('Temperature Over Time')
    plt.grid(True)
    plt.legend()

    # Plot Humidity
    plt.subplot(3, 1, 2)
    plt.plot(dates, humidity, color='b', marker='x', linestyle='-', label='Humidity')
    plt.xlabel('Index')
    plt.ylabel('Humidity (%)')
    plt.title('Humidity Over Time')
    plt.grid(True)
    plt.legend()

    # Plot Wind Speed
    plt.subplot(3, 1, 3)
    plt.plot(dates, wind_speed, color='g', marker='s', linestyle='-', label='Wind Speed')
    plt.xlabel('Index')
    plt.ylabel('Wind Speed (km/h)')
    plt.title('Wind Speed Over Time')
    plt.grid(True)
    plt.legend()

    # Adjust layout to avoid overlap
    plt.tight_layout()
    plt.show()

# Main program
filename = 'weather_data.csv'  # Nama file CSV
data = csv_to_numpy(filename)  # Membaca CSV ke dalam NumPy array

# Analisis data
dates, temperature, humidity, wind_speed = analyze_data(data)

# Visualisasi data
plot_data(dates, temperature, humidity, wind_speed)
```
